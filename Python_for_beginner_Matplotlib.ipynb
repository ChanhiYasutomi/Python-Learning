```python
# 08. 凡例
# 以下のコードを実行したグラフに対しsin,cosという判例をつけたグラフを作成してください

# x = np.linspace(0,2*np.pi, 500)
# y1 = np.sin(x)
# y2 = np.cos(x)

# plt.plot(x, y1)
# plt.plot(x, y2)

# plt.ylim(-2,2)

x = np.linspace(0,2*np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)

print(x)
print(y1)
print(y2)

# sin,cosという判例をつけたグラフを作成してください
plt.plot(x, y1,label="sin")
plt.plot(x, y2,label="cos")
plt.ylim(-2,2)

#　判例()をつけたい場合 - plt.legend()
# legend = 伝説、判例

plt.legend()

# plt.legend(loc=2) - 判例を左側に表示したい場合
# loc = location
# (loc=2) - 文字列でもコードでも大丈夫
```

```python
# 15. 3次元グラフ
# 次のデータを使って3次元に点をプロットしてみてください。
# こちらのデータは z軸真上方向に螺旋のような動きをしながら伸びていくものとなっております。

# theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
# z = np.linspace(-2,2,100) 
# r = z**2+1
# x = r * np.sin(theta)
# y = r * np.cos(theta)

# 3次元グラフに必要モジュール

from mpl_toolkits.mplot3d import Axes3D

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2,2,100) 
r = z**2+1
x = r * np.sin(theta)
y = r * np.cos(theta)

fig = plt.figure()

# fig.gca(projection="3d") - 3dの設定
ax = fig.gca(projection="3d")
print(ax)

# 3dでプロットする
# ax.plot(x軸,y軸,z軸,label="3D Curve")
ax.plot(x,y,z,label="3D Curve")
ax.legend()
plt.show()

# 2次元で見たい場合
plt.plot(x,y)
```

```python
# 16. 円グラフ
# 以下にに用意したデータを使って円グラフを作成してください
 
# x = np.array([100,200,300,400,500])
# labels = ["Apple","Banana","Orange","Grape","Strawberry"]
# colors = ["red","yellow","orange","purple","pink"]

x = np.array([100,200,300,400,500])
labels = ["Apple","Banana","Orange","Grape","Strawberry"]
colors = ["red","yellow","orange","purple","pink"]

# plt.pie() - 円グラフの作成
plt.pie(x,labels=labels,colors=colors)

# 右回りにしたい場合
plt.pie(x,labels=labels,colors=colors,counterclock=False)

# 右回りにしたいかつ、頂点から始めたい場合
plt.pie(x,labels=labels,colors=colors,counterclock=False,startangle=90)
```

```python
# 19. 演習問題2
# 次のデータを使って以下の図のようなグラフを作成してください

# label = "Chips","Hotdogs","Cookies","Chocolates"
# sizes = [15,30,45,10]

# なお、円グラフの特徴は以下になります
# 円グラフ(Hogsのみ半径10%分のオフカット)
# 影あり
# 少数点以下1桁%まで表示する
# 0時の位置から反時計回りに始まる

labels = "Chips","Hotdogs","Cookies","Chocolates"
sizes = [15,30,45,10]
explode = (0,0.1,0,0) #(0=chips,0.1=Hotdogs,0=Cookies,0=Chocolates) 
# 0.1 - ぐらい離れる

# autopct="%1.1f" - %1.1f 小数点以下1%を表示する
# shadow=True - 影
plt.pie(sizes,explode=explode,labels=labels,autopct="%1.1f%%",shadow=True,startangle=90)

#縦横比を揃える - plt.axis("equal")
plt.axis("equal")
plt.show()
```

```python
# 20. 演習問題3
# 次の3つの異なるデータを使って以下の図のような積み上げタイプのヒストグラムを作成してください。
# ただしビンの数は10としています

# mu1,sigma1 = 100,15
# mu2,sigma2 = 90,20
# mu3,sigma3 = 110,10

# x1 = mu1 + sigma1 * np.random.randn(100)
# x2 = mu2 + sigma2 * np.random.randn(100)
# x3 = mu3 + sigma3 * np.random.randn(100)

mu1,sigma1 = 100,15
mu2,sigma2 = 90,20
mu3,sigma3 = 110,10

x1 = mu1 + sigma1 * np.random.randn(10000)
x2 = mu2 + sigma2 * np.random.randn(10000)
x3 = mu3 + sigma3 * np.random.randn(10000)

print(x1)
print(x2)
print(x3)

# stacked=True - 積み上げ
# plt.hist([x1,x2,x3],bins=10,color=["yellow","skyblue","blue"],label=["x1","x2","x3"],stacked=True)

# より見やすくする
# 0-1正規化する - normed = True
# plt.hist([x1,x2,x3], bins=10, normed=True, color=["yellow","skyblue","blue"],label=["x1","x2","x3"],stacked=True)
# normed - では無理だったためdensityで代替した
plt.hist([x1,x2,x3], bins=10, density=True, color=["yellow","skyblue","blue"],label=["x1","x2","x3"],stacked=True)
plt.title("histgram")
plt.xlabel("x")
plt.ylabel("freq")
plt.xlim(0,180)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
